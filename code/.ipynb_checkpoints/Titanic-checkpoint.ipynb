```python
def exp_int_point (train):
    int_cols = train[train.dtypes[train.dtypes == 'int'].index]
    shapedf = int_cols.shape
    nbcol = shapedf[1]
    nbrow = shapedf[0]
    width = 3
    if nbcol < width:
        fig,ax1 = plt.subplots(1,width,figsize=(5*width,5*(nbcol/3+1)))
        fig,ax2 = plt.subplots(1,width,figsize=(5*width,5*(nbcol/3+1)))
        for i,col in enumerate(list(int_cols.columns.values)):
            sns.distplot(train[col].dropna(),ax=ax2[i%width]) 
            ax1[i%width].plot(np.arange(nbrow ),int_cols[col].sort_values())
            ax1[i%width].set_title(col)
    else:
        fig1,ax1 = plt.subplots(nbcol/width+1,width,figsize=(5*(nbcol/3+1),7))
        for i,col in enumerate(list(int_cols.columns.values)):
            ax1[i/width,i%width].plot(np.arange(nbrow ),int_cols[col].sort_values())
            ax1[i/width,i%width].set_title(col)
        fig2,ax2 = plt.subplots(nbcol/width+1,width,figsize=(5*width,5*(nbcol/3+1)))
        for i,col in enumerate(list(int_cols.columns.values)):
            sns.distplot(train[col],ax=ax2[i/width,i%width])
    fig1.savefig('Summary_Dataexplore_col_int_point.pdf')
    fig2.savefig('Summary_Dataexplore_col_int_dist.pdf')
#exp_int_point (train)
```

```python
def exp_float_point (train):
    flt_cols = train[train.dtypes[train.dtypes == 'float'].index]
    shapedf = flt_cols.shape
    nbcol = shapedf[1]
    nbrow = shapedf[0]
    width = 3
    if nbcol < width:
        fig,ax = plt.subplots(1,width,figsize=(5*width,5*(nbcol/3+1)))
        for i,col in enumerate(list(flt_cols.columns.values)):
            sns.distplot(train[col].dropna(),ax=ax[i%width]) 
    else:
        fig,ax = plt.subplots(nbcol/width+1,width,figsize=(5*(nbcol/3+1),5))
        for i,col in enumerate(list(flt_cols.columns.values)):
            sns.distplot(train[col].dropna(),ax=ax[i/width,i%width]) 
    fig.savefig('Summary_Dataexplore_col_float_dist.pdf')
#exp_float_point (train)
```

```python
def exp_cate (train):
    cate= train[train.dtypes[train.dtypes == 'object'].index]
    a = cate.isnull().sum()
    b = cate.apply(lambda x:( x.value_counts()>0).sum())
    cate_sum=pd.concat([a,b],axis=1)
    cate_sum.columns=['miss_ct','unq_ct']
    cate_toplot_cols = train[cate_sum['unq_ct'][(cate_sum['unq_ct']<10)].index]
    shapedf = cate_toplot_cols.shape
    nbcol = shapedf[1]
    nbrow = shapedf[0]
    width = 3
    if nbcol < width:
        fig,ax = plt.subplots(1,width,figsize=(5*width,5*(nbcol/3+1)))
        for i,col in enumerate(list(cate_toplot_cols.columns.values)):
            sns.countplot(train[col].dropna(),ax=ax[i%width]) 
    else:
        fig,ax = plt.subplots(nbcol/width+1,width,figsize=(5*(nbcol/3+1),5))
        for i,col in enumerate(list(cate_toplot_cols.columns.values)):
            sns.distplot(train[col].dropna(),ax=ax[i/width,i%width]) 
    return cate_sum
cate_sum = exp_cate (train)
```
